```python
import pandas as pd
raw_data = pd.read_csv("collected_data.csv")
pd.options.display.max_columns = 99
raw_data
```

# Cleaning data
Correlation matrix was not displaying all columns as the crime columns were of string type

```python
pd.options.display.max_rows = 99
raw_data.dtypes
```

```python
# Replace letter 'D' in column "Total manufacturers shipments ($1000) 2012"
raw_data.loc[raw_data["Total manufacturers shipments ($1000) 2012"] == "D", "Total manufacturers shipments ($1000) 2012"] = 0
raw_data
```

```python
# separating object and numeric col
object_col = raw_data[['State', 'Code', 'Electoral Result']]
numeric_col = raw_data.drop(columns = ['State', 'Code', 'Electoral Result'])
```

```python
# removing commas from strings in numeric columns and converting them to numeric dtype
numeric_col.replace(',','', regex=True, inplace=True)
c = numeric_col.select_dtypes(object).columns
numeric_col[c] = numeric_col[c].apply(pd.to_numeric,errors='coerce')
numeric_col
```

```python
# combining both dataframes
frames = [object_col, numeric_col]
df = pd.concat(frames, axis = 1)
df
```

# Correlation

```python
corr = df.corr()
pd.options.display.max_rows = 99
corr
```
