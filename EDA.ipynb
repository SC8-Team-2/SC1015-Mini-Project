```python
import pandas as pd
import numpy as np
raw_data = pd.read_csv("collected_data.csv")
pd.options.display.max_columns = 99
raw_data
```

# Cleaning data
Correlation matrix was not displaying all columns as the crime columns were of string type

```python
pd.options.display.max_rows = 99
raw_data.dtypes
```

```python
# Replace letter 'D' in column "Total manufacturers shipments ($1000) 2012"
raw_data.loc[raw_data["Total manufacturers shipments ($1000) 2012"] == "D", "Total manufacturers shipments ($1000) 2012"] = 0
raw_data
```

```python
# separating object and numeric col
object_col = raw_data[['State', 'Code', 'Electoral Result']]
numeric_col = raw_data.drop(columns = ['State', 'Code', 'Electoral Result'])
```

```python
# removing commas from strings in numeric columns and converting them to numeric dtype
numeric_col.replace(',','', regex=True, inplace=True)
c = numeric_col.select_dtypes(object).columns
numeric_col[c] = numeric_col[c].apply(pd.to_numeric,errors='coerce')

numeric_col
```

```python
# combining both dataframes
frames = [object_col, numeric_col]
df = pd.concat(frames, axis = 1)
df['Electoral Result'] = np.where(df['Electoral Result'].eq('R'), 1, 0)
df
```

```python
crimes = [
    "Total all classes", 
    "Violent crime", 
    "Property crime", 
    "Murder and nonnegligent manslaughter", 
    "Rape", 
    "Robbery", 
    "Aggravated assault", 
    "Burglary", 
    "Larceny- theft", 
    "Motor vehicle theft",
    "Arson",
    "Other assaults",
    "Forgery and counterfeiting",
    "Fraud",
    "Embezzlement",
    "Stolen property",
    "Vandalism",
    "Weapons",
    "Prostitution",
    "Sex offenses (except rape and prostitution)",
    "Drug abuse violations",
    "Gambling",
    "Offenses against the family and children",
    "Driving under the influence",
    "Liquor laws",
    "Drunkenness",
    "Disorderly conduct",
    "Vagrancy",
    "All other offenses (except traffic)",
    "Suspicion",
    "Curfew and loitering law violations"
]
```

```python
# rebase values
multiplier = 1000000 # one million
for index, row in df.iterrows():
    pop = df.at[index, "Population Estimate"]
    df.at[index, "Veterans"] = float(multiplier * df.at[index, "Veterans"] / pop)
    for crime in crimes:
        df.at[index, str(crime) + " arrests"] = float(multiplier * df.at[index, str(crime) + " arrests"] / pop)        
        
df
```

# Correlation

```python
corr = df.corr().unstack().sort_values().drop_duplicates()
pd.set_option('display.max_rows', None)
pd.DataFrame(corr)
```
