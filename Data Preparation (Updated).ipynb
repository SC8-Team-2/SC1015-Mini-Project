# Preparation of Suicide Rates Data

```python
import pandas as pd
import numpy as np
csv_data = pd.read_csv("Suicide Dataset.csv")
csv_data
```

```python
year_count = pd.DataFrame(csv_data['Year'].value_counts())
year_count
```

## Select rows from year 2019

```python
year_2017 = csv_data[csv_data['Year'] == 2017]
year_2017.reset_index(inplace = True)
```

```python
year_2017
```

## Filtering country and suicide columns

```python
focus = year_2017[['Country', 'Crude suicide rates (per 100 000 population) (Both sexes)']]
for index in range(len(year_2017)):
    # filter cell value with '.at' method
    cell_value = focus.at[index, 'Crude suicide rates (per 100 000 population) (Both sexes)']
    # cast cell value from series to string, split into a list and index the first value
    focus.at[index, 'Crude suicide rates (per 100 000 population) (Both sexes)'] = str(cell_value.split()[0])

focus
```

# Preparation of Country Statistics Data

```python
country_stats = pd.read_csv('country_profile_variables.csv')
print(country_stats.shape)
display(country_stats)
```

## List of Variables

```python
for col in country_stats.columns:
    print(col)
```

## Filtering Variables
We filtered the variables such that we would only be left with Numerical Data

```python
country_stats.drop(columns=['Labour force participation (female/male pop. %)'], inplace=True)
country_stats.drop(columns=['Life expectancy at birth (females/males, years)'], inplace=True)
country_stats.drop(columns=['Population age distribution (0-14 / 60+ years, %)'], inplace=True)
country_stats.drop(columns=['International migrant stock (000/% of total pop.)'], inplace=True)
country_stats.drop(columns=['Education: Primary gross enrol. ratio (f/m per 100 pop.)'], inplace=True)
country_stats.drop(columns=['Education: Secondary gross enrol. ratio (f/m per 100 pop.)'], inplace=True)
country_stats.drop(columns=['Education: Tertiary gross enrol. ratio (f/m per 100 pop.)'], inplace=True)
country_stats.drop(columns=['Energy supply per capita (Gigajoules)'], inplace=True)
country_stats.drop(columns=['Forested area (% of land area)'], inplace=True)
country_stats.drop(columns=['Pop. using improved drinking water (urban/rural, %)'], inplace=True)
country_stats.drop(columns=['Net Official Development Assist. received (% of GNI)'], inplace=True)

display(country_stats)
```

## Filtering Countries
We filtered out the countries from the Country Statistics Data which were not found in Suicide Rates data, and vice versa

```python
for index, row in country_stats.iterrows():
    if row['country'] not in focus['Country'].tolist():
        print(f'{row["country"]} has been removed')
        country_stats.drop(index, inplace=True)
print(country_stats.shape)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
country_stats = country_stats.reset_index(drop=True)
display(country_stats)
```

```python
temp = country_stats.loc[:, 'country'].tolist()
for index, row in focus.iterrows():
    if row['Country'] not in temp:
        print(f'{row["Country"]} has been removed')
        focus.drop(index, inplace=True)
print(focus.shape)
focus = focus.reset_index(drop = True)
display(focus)
```

```python
# Checking if both datasets have same countries
country_stats['country'].isin(focus['Country']).value_counts()
```

## Cleaning Data

### Correction of Invalid Data

```python
country_stats.replace('~0.0', 0.0, inplace = True)
country_stats.replace('-~0.0', 0.0, inplace = True)
country_stats.replace('...', np.nan, inplace = True)
country_stats.replace(-99, np.nan, inplace = True)
```

### Correction of Invalid Datatypes

```python
for col in list(country_stats.columns.values):
    if col != 'country' and col != 'Region':
        country_stats[col] = country_stats[col].astype(float)
country_stats.dtypes
```

## Combining Datasets

```python
final = pd.concat([country_stats, focus], axis = 1).reindex(country_stats.index)
final.drop(['Country'], axis=1)
```

```python
final.to_csv('final_data.csv', index=False)
```
