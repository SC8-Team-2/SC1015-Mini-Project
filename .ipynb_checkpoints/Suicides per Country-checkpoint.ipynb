```python
import pandas as pd
```

```python
csv_data = pd.read_csv("final_table2.csv")
csv_data
#another comment
```

# Year distribution

```python
years = csv_data['year']
years.value_counts()
```

# Filtering rows from year 2014

```python
csv_data = csv_data[csv_data['year'] == 2014]
csv_data
```

# Filtering country and suicide columns


```python
focus = csv_data[['country', 'suicides/100k pop']]
focus
```

# Table of Country : Suicides per 100k population

```python
aggregation_functions = {'suicides/100k pop': 'sum'}
final = focus.groupby(focus['country']).aggregate(aggregation_functions)
```

```python
final.describe()
```

## Alphabetical Order

```python
final
```

## Descending Order

```python
final.sort_values(by = ['suicides/100k pop'], ascending = False)
```
