```python
import pandas as pd
csv_data = pd.read_csv("Suicide Dataset.csv")
csv_data
```

```python
year_count = pd.DataFrame(csv_data['Year'].value_counts())
year_count
```

# Select rows from year 2019

```python
year_2019 = csv_data[csv_data['Year'] == 2019]
year_2019.reset_index(inplace = True)
```

```python
year_2019
```

# Filtering country and suicide columns

```python
focus = year_2019[['Country', 'Crude suicide rates (per 100 000 population) (Both sexes)']]
for index in range(len(year_2019)):
    # filter cell value with '.at' method
    cell_value = focus.at[index, 'Crude suicide rates (per 100 000 population) (Both sexes)']
    # cast cell value from series to string, split into a list and index the first value
    focus.at[index, 'Crude suicide rates (per 100 000 population) (Both sexes)'] = str(cell_value.split()[0])

focus
```
